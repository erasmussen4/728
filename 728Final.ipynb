**From Section 4.1 in Kolb and Turner's *The Early Universe*:**

When a non-relativistic species (with mass number $A$ and charge $Z$) is in kinetic equilibrium, the number density is given by
\begin{equation}
    n_A = g_A \left( \frac{m_A T}{2\pi} \right)^{3/2}\exp\left(\frac{\mu_A-m_A}{T}\right)~.
\end{equation}
When these reactions happen quickly compared to the rate of expansion of the universe, the species is also in chemical equiliubrium. In this case, we can write the chemical potential in terms of those for the proton and neutron
\begin{equation}
    \mu_A = Z\mu_p +(A-Z)\mu_n~.
\end{equation}
From this, we can rewrite the exponential in Eqn. (1) as 
\begin{align}
    \exp\left(\frac{\mu_A}{T}\right) 
    &= \exp\left[\frac{Z\mu_p+(A-Z)\mu_n}{T}\right] \nonumber\\
    &= 2^{-A}n_p^Z n_n^{A-Z}\left(\frac{2\pi}{m_N T}\right)^{3A/2}\exp\left[\frac{Zm_p+(A-Z)m_n}{T}\right]~,
\end{align}
where $m_N$ is the nucleon mass, which we have used as a stand-in for all the the mass differences bewteen $m_n$, $m_p$, and $m_A/A$.

Recall that the binding energy is given by
\begin{equation}
    B_A = Zm_p + (A-Z)m_n - m_A~.
\end{equation}
Using this and our result from Eqn. (3), we can write the number density as
\begin{equation}
    n_A = g_A A^{3/2}2^{-A}n_p^Z n_n^{A-Z}
    \left(\frac{2\pi}{m_N T}\right)^{3(A-1)/2}
    \exp\left(\frac{B_A}{T}\right)~.
\end{equation}
See the Table below for a list of binding energies. 

$$
\begin{array}{c|c|c}
    ^AZ & B_A \text{(MeV)} & g_A \\
    \hline
    ^2\text{H} & 2.22 & 3 \\
    ^3\text{H} & 8.48 & 2 \\
    ^3\text{He} & 7.72 & 2 \\
    ^4\text{He} & 28.3 & 1 \\
    ^7\text{Li} & 39.25 & 4 \\
    ^7\text{Be} & 37.6 & 4
\end{array}
$$

Because the expansion of the universe decreases the number density, we will instead use the total nucleon density 
$$n_N = n_n + n_p +\sum_i (An_A)_i$$
as a fiducial quantity and consider the mass fraction
\begin{equation}
    X_A = \frac{An_A}{n_N}\quad\text{with}\quad \sum_i X_i = 1~.
\end{equation}
From this definition, the mass fraction in nuclear statistical equilibrium is found to be
\begin{equation}
    X_A = g_A\left[\zeta^{A-1}(3)2^{(3A-5)/2}\pi^{(1-A)/2}\right]A^{5/2}
    \left(\frac{T}{m_N}\right)^{3(A-1)/2}\eta^{A-1}X_p^Z X_n^{A-Z}\exp\left(\frac{B_A}{T}\right)~,
\end{equation}
where $\eta$ is the ratio of baryons to photons today, given by
\begin{equation}
    \eta\equiv\frac{n_N}{n_\gamma}=2.68\times 10^{-8} (\Omega_B h^2)~.
\end{equation}

**Other helpful hints:**

We will want to separate the nuclear reaction rates, i.e. vs. Boltzmann evolution eqns, vs. temperature dependencies, etc. It may be helpful to organize the rates as
$$\frac{d}{dt}\left(\frac{X_i}{A_i}\right) = \sum_{j=1}^{40}\Delta_{ij}f_i g_j~,$$
where we sum over all reactions that create and destroy the light elements. The $\Delta_{ij}$ specifies the quantity of a reacting species that is created or destroyed in a reaction. The $g_j$ depends on the number of reacting species, i.e.
$$g_j = \frac{X_k}{A_k}
\quad\text{or}\quad
\frac{X_k}{A_k}\frac{X_\ell}{A_\ell}
\quad\text{or}\quad
\frac{X_k}{A_k}\frac{X_\ell}{A_\ell}\frac{X_m}{A_m}~,$$
where $k$, $\ell$, and $m$ run over the nuclei: $p$, $n$, D, T, $^3\mathrm{He}$, $^4\mathrm{He}$, $^7\mathrm{Li}$, and $^7\mathrm{Be}$. Lastly, $f_i$ are coefficients which are functions of: 
- the baryon density, $\rho_B$,
- temperature, $T$,
- and a number of free coefficients found in the paper from Wagoner *et al.*

(We will need to determine the functional form of $f_i$ ourselves.)

where 
$$f_i(\rho_b,T_9) = \lambda_\gamma(j)\quad\text{or}\quad
f_i(\rho_b,T_9) = [jk]\quad\text{or}\quad

We can convert our time derivative to a derivative in $x_i\equiv \frac{Q_i}{T}$, using
$$\frac{dX_i}{dt} = \frac{dx_i}{dt}\frac{dX_i}{dx_i} =\dots

$$g_* = \sum_\text{bosons}g_i\left(\frac{T_i}{T}\right)^4 + \frac{7}{8}\sum_\text{fermions}g_i\left(\frac{T_i}{T}\right)^4~,$$
with $g_i=g_\text{particles}\cdot g_\text{spins}\cdot g_\text{colors}$.
